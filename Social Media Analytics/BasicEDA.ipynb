```python
from textblob import TextBlob
```

```python
comment = 'Enjoying a beautiful day at the park!'
pol = TextBlob(comment).sentiment.polarity
```

```python
pol
```

```python
def give_sentiment_score(comment):
  return TextBlob(comment).sentiment.polarity
```

```python
df['sentiment_score'] = df['Text'].apply(give_sentiment_score)
```

```python
def give_sentiment(score):
  if score>0:
    return 'positive'
  elif score<0:
    return 'negative'
  else:
    return 'neutral'
```

```python
df['sentiment_score'] = df['sentiment_score'].apply(give_sentiment)
```

```python
df.head()
```

```python
pd.crosstab(df['sentiment_score'],df['Platform']).plot(kind='bar',stacked=True)
```

```python
import re
def hash_tag_remover(text):
  for i in (re.findall(r'#(\w+)',text)):
    return i
```

```python
hash_tag_remover(' Excited about the upcoming weekend getaway! #Morning')
```

```python
df['processed_tags'] = df['Hashtags'].apply(hash_tag_remover)
```

```python
df['processed_tags'][:100].value_counts().plot(kind='bar',figsize=(20,4))
```
