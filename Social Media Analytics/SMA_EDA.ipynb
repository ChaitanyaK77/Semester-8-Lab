```python
def assign_sentiment(score):
  if score>0:
    return 'positive'
  elif score<0:
    return 'negative'
  else:
    return 'neutral'
```

```python
df1['sentiment_score'] = df1['sentiment_score'].apply(assign_sentiment)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
comments = df1['text'].astype(str).str.lower().values
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
vectorizer = TfidfVectorizer(stop_words=stop_words, max_df=0.95, min_df=2)
X = vectorizer.fit_transform(comments)
```

```python
print(X)
```

```python
lda = LatentDirichletAllocation(n_components=5,random_state=42)
lda.fit(X)
```

```python
words = vectorizer.get_feature_names_out()
topics = lda.components_
```

```python
for i in range(3):
  top_words = words[topics[i].argsort()[:5]]
  print(top_words)
```
